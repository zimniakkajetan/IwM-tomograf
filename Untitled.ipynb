```python
import numpy as np
from matplotlib import pyplot as plt
from skimage.color import rgb2gray
from skimage import data, io
    
def bresenhamLine(x1, y1, x2, y2):

    line = []
    if (x1 <= x2):
        xi = 1 
    else:
        xi = -1
    if(y1 <= y2):
        yi = 1
    else:
        yi = -1

    dx = abs(x2 - x1)
    dy = abs(y2 - y1)

    x = x1
    y = y1
    line.append([x, y])
    if (dx >= dy):
        ai = (dy-dx) * 2
        bi = dy * 2
        d = bi - dx
        while(x != x2):
            if(d >= 0):
                x += xi
                y += yi
                d += ai
            else:
                d += bi
                x += xi
            line.append([x,y])
    else:
        ai = (dx - dy) * 2
        bi = dx *2
        d = bi - dy
        while( y != y2):
            if(d >= 0):
                x += xi
                y +=yi
                d += ai
            else:
                d += bi
                y += yi
            line.append([x,y])
            
    return line


def makeSinogram(pic):
    numbersOfDetectors = 50
    alpha = 3
    rozpietosc = 90
    pic_size = len(pic[0])
    
    r = pic_size

    sinogram = []

    for i in range(0, 360, alpha):
        sinogram.append([])
        katRadiany = i * np.pi / 180
        for detector in range(0, numbersOfDetectors):
            x0 = r * np.cos(katRadiany)
            y0 = r * np.sin(katRadiany)

            x1 = r * np.cos(katRadiany + np.pi - ((rozpietosc / 2)*np.pi/180) + ((detector * (rozpietosc / (numbersOfDetectors - 1)))*np.pi/180))
            y1 = r * np.sin(katRadiany + np.pi - ((rozpietosc / 2)*np.pi/180) + ((detector * (rozpietosc / (numbersOfDetectors - 1)))*np.pi/180))

            x0 = int(x0) + np.floor(pic_size / 2)
            x1 = int(x1) + np.floor(pic_size / 2)
            y0 = int(y0) + np.floor(pic_size / 2)
            y1 = int(y1) + np.floor(pic_size / 2)

            line = bresenhamLine(x0, y0, x1, y1)

            pixel = np.float(0)
            pixLicz = int(0)
            for p in line:
                if p[0] >= 0 and p[1] >= 0 and p[0] < len(picture) and p[1] < len(picture):
                    pixel += float(pic[int(p[0]),int(p[1])])
                    pixLicz +=1

            sinogram[-1].append(pixel/pixLicz)

    return sinogram



if __name__ == "__main__":
   
    #picture = np.zeros([200,200])
    picture = rgb2gray(io.imread("foto1/Shepp_logan.jpg"))
    
    plt.figure()
    plt.title("Zdjecie wejsciowe")
    plt.imshow(picture, cmap='gray')
    plt.show()
    
    sinogramP = makeSinogram(picture)
    plt.title("Sinogram")
    plt.imshow(sinogramP,cmap='gray')
    #print((sinogramP))
    plt.show()
    
    
```
